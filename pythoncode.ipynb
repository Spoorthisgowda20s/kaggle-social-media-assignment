```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("Students Social Media Addiction.csv")

# Rename columns for convenience
df.rename(columns={
    'Avg_Daily_Usage_Hours': 'Daily_Hours',
    'Sleep_Hours_Per_Night': 'Sleep_Hours',
    'Affects_Academic_Performance': 'Affects_Academics',
    'Conflicts_Over_Social_Media': 'Conflicts',
    'Addicted_Score': 'Addiction_Score',
    'Mental_Health_Score': 'Mental_Health'
}, inplace=True)

# Set theme
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 1. Most Used Platform
plt.subplot(3, 1, 1)
platform_counts = df['Most_Used_Platform'].value_counts().head(10)
sns.barplot(x=platform_counts.values, y=platform_counts.index, palette="viridis")
plt.title("Top 10 Most Used Social Media Platforms")
plt.xlabel("Number of Students")
plt.ylabel("Platform")

# 2. Addiction Score vs. Daily Usage Hours
plt.subplot(3, 1, 2)
sns.scatterplot(data=df, x="Daily_Hours", y="Addiction_Score", hue="Gender", palette="Set1", alpha=0.7)
plt.title("Addiction Score vs Daily Social Media Usage")
plt.xlabel("Daily Usage Hours")
plt.ylabel("Addiction Score")

# 3. Effect on Academics by Gender
plt.subplot(3, 1, 3)
sns.countplot(data=df, x="Affects_Academics", hue="Gender", palette="pastel")
plt.title("Does Social Media Affect Academics? (By Gender)")
plt.xlabel("Affects Academics")
plt.ylabel("Number of Students")

plt.tight_layout()
plt.show()
```
